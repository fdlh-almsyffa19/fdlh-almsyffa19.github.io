```python
import numpy as np
#Algoritma Enkripsi Hill Cipher
#Input;pesan,kunci
#ouput: ciphertext
#langkah
#1. Ubah kunci ke dalam matrix ukuran nxn
#2. Ubah pesan ke dalam matriks yang ukurannya sama dengan ukuran matiks
#3. Ubah kunci dan pesan ke dalam angka 0-25
#4. Hitung : E= Pi x K


def hillecrypt(pesan,kunci):
  n=len(kunci)
  ciphertext= ""
  while len(pesan)%n!= 0:
    pesan+= 'X'

  for i in range(0,len(pesan),n):
    blok= pesan[i:i+n]
    blok_angka=[ord(char) % 65 for char in blok]
    encrypt_angka= np.dot(kunci,blok_angka)%26
    ciphertext_blok=''.join(chr(num+65) for num in encrypt_angka)
    ciphertext +=ciphertext_blok
  return ciphertext

pesan ="HELLO"
kunci=np.array([[1,0,1],[3,1,2],[2,1,0]])

ciphertext=hillecrypt(pesan,kunci)
print(ciphertext)
```

```python
from re import M
#invers modulo

import numpy as np
def inversmod(a,m):
  m0,x0,x1=m,0,1
  if m==1:
    return 0

  a = (a % m + m) % m
  while a>1:
    q=a//m
    m, a=a%m,m
    x0, x1=x1-q*x0, x0
  if x1<0:
    x1 += m0
  return x1

d=inversmod(5,3)
print(d)
```

```python
import numpy as np

def matrix_mod_inv(matrix, modulus):
    det = round(np.linalg.det(matrix)) # Calculate determinant
    det_inv = inversmod(det, modulus) # Find modular inverse of determinant (using existing inversmod)

    matrix_2d = np.array(matrix, dtype=int)

    adjugate = np.zeros(matrix_2d.shape, dtype=int)

    # Calculate adjugate matrix (transpose of cofactor matrix)
    for i in range(matrix_2d.shape[0]):
        for j in range(matrix_2d.shape[1]):
            # Create minor matrix
            minor_matrix = np.delete(np.delete(matrix_2d, i, axis=0), j, axis=1)
            cofactor = round(np.linalg.det(minor_matrix))
            adjugate[j, i] = ((-1)**(i+j) * cofactor) % modulus # (j, i) for transpose

    # Calculate inverse matrix
    inv_matrix = (det_inv * adjugate) % modulus
    return inv_matrix

def hilldecrypt(ciphertext, key):
    n = len(key)
    plaintext = ""

    key_inv = matrix_mod_inv(key, 26)

    for i in range(0, len(ciphertext), n):
        blok = ciphertext[i:i+n]
        blok_angka = [ord(char) % 65 for char in blok]

        decrypt_angka = np.dot(key_inv, blok_angka) % 26

        plaintext_blok = ''.join(chr(num + 65) for num in decrypt_angka)
        plaintext += plaintext_blok
    return plaintext

# Example usage with the previous key and ciphertext
pesan = "HELLO"
kunci = np.array([[1,0,1],[3,1,2],[2,1,0]]) # The key used in the corrected encryption
ciphertext = "SVSIPK" # The result from the corrected encryption

decrypted_message = hilldecrypt(ciphertext, kunci)
print(f"Ciphertext: {ciphertext}")
print(f"Decrypted: {decrypted_message}")
```
