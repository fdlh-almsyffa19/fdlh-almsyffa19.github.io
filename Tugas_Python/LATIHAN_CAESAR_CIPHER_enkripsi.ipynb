```python
#================================================
# Algoritma Caesar Cipher untuk proses Enkripsi
#===============================================
def enkripsi(pesan,k):
  ciphertext=""
  l=len(pesan)
  for i in range(l):
    char = pesan[i] #merubah pesan jadi individual huruf
    s = (ord(char) + k - 65)%26
    #merubah setiap huruf menjadi angka dengan fungsi ord
    ciphertext += chr(s +65)
  return ciphertext

pesan ="INDONESIA"
k = 3
ciphertextpesan=enkripsi(pesan,k)
print('pesan:',pesan)
print('kunci:',k)
print('ciphertext:',ciphertextpesan)
```
