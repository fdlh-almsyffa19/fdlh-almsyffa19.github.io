```python
#Algoritma Deskripsi Ceesar Cipher#
#input:ciphertext,kunci
#output:pesan
#langkah:
#========================================================================
def decrypt(ciphertext,kunci):
  pesan=""
  l=len(ciphertext)
  for i in range(l):
    char =ciphertext[i]
    s =(ord(char) - kunci - 65)%26
    pesan += chr(s + 65)
  return pesan

ciphertext ="LQGRQHVLD"
kunci = 3
dekripsipesan = decrypt(ciphertext,kunci)
print('ciphertext:',ciphertext)
print('kunci:',kunci)
print('pesan:',dekripsipesan)
```
